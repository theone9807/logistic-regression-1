```python

import statsmodels.api as sm  #used for data 
import numpy as np 
import seaborn as sns
import pandas as pd
import math
from pandas import Series as se, DataFrame as df
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
df2 = sm.datasets.fair.load_pandas().data
df2 = df(df2)
df2.head()
```

```python
def check_affair(x):
    
    if x != 0:
        return 1
    return 0

df2['had_affairs'] = df2['affairs'].apply(check_affair)  #to directely apply the func. on data set
```

```python
df2.head()
```

```python
df2.groupby('had_affairs').mean()
```
